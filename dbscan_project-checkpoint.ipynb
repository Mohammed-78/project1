```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime=pd.read_csv('en.csv')
crime
```

```python
crime.info()
```

```python
crime
```

```python
crime.drop(['State/UT'],axis=1,inplace=True)
crime
```

```python
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
dbscan.labels_
```

```python
crime['clusters']=dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clusters'],crime['Rate of cognizable crimes'], c=dbscan.labels_)
```
